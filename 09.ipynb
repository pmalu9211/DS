```python
import pandas as pd

df = pd.read_csv('/content/dataset_Facebook.csv', sep=';')
df.head(10)
```

```python
# 2a. Subset for Photo posts
#     - Boolean indexing: pick rows where df['Type'] equals 'Photo'
photo_df = df[df['Type'] == 'Photo']
print("=== Photo Subset ===")
photo_df.head(10)
```

```python
# 2b. Subset for Status posts
status_df = df[df["Type"] == "Status"]
print("=== Status Subset ===")
status_df.head(10)
```

```python
# 3. Merge the two subsets back together
#    - pd.concat stacks them vertically
#    - ignore_index=True resets the row numbers
merged_df = pd.concat([photo_df, status_df], ignore_index=True)    #IMP
merged_df.head()
# merged_df.tail()
```

```python
# 4. Sort the entire DataFrame by 'Page total likes'
#    - ascending=True gives lowest → highest
sorted_df = df.sort_values(by='Page total likes', ascending=True)    #IMP
print("=== Sorted by Page total likes ===")
sorted_df[['Page total likes', 'Type']].head()
```

```python

# 5. Transpose the DataFrame
#    - .T flips rows ↔ columns
transposed_df = df.head().T
transposed_df
```

```python
# 6. Melt into “long” format
#    - id_vars stay as-is; value_vars get unpivoted into two columns
melted_df = pd.melt( #IMP
    df,
    id_vars=['Page total likes', 'Type'],
    value_vars=['comment', 'like', 'share'],
    var_name='Interaction',
    value_name='Count'
)
melted_df.shape
```

```python
# 7. Cast back to “wide” format
#    - pivot_table handles any duplicate (page‑type) combos
#    - aggfunc='first' just takes the first value if there’s more than one
wide_df = melted_df.pivot_table( #IMP
    index=['Page total likes', 'Type'],
    columns='Interaction',
    values='Count',
    aggfunc='first'
).reset_index()

wide_df.shape
```
